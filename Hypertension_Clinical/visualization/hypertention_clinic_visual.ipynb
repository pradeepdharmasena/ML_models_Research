```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt
import sklearn.metrics as sm
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import MinMaxScaler
from scipy import stats
import seaborn as sb
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv("../data/hypertension_clinical.csv")
df = df.drop_duplicates()
print(df.head())
```

```python
cols = df.columns
print(len(cols))
print(cols)
```

```python
fig, axes = plt.subplots(6,2,figsize=(20,30))
i = 0
j = 0
for col in cols[:]:
    sb.histplot(data=df, x=df[col],  ax=axes[i,j])
    plt.xticks(ticks=df[col].unique())
    
    j = j + 1
    
    if j >= 2:
        j = 0
        if i < 5:
            i = i + 1

plt.show()
```

```python
fig, axes = plt.subplots(6,2,figsize=(20,30))
i = 0
j = 0
for col in cols[:]:
    sb.boxplot(x=df[col],ax=axes[i][j])
    plt.xticks(ticks=df[col].unique())
    
    j = j + 1
    
    if j >= 2:
        j = 0
        if i < 5:
            i = i + 1

plt.show()
```

```python
fig, axes = plt.subplots(6,2,figsize=(20,30))
i = 0
j = 0
for col in cols[:]:
    sb.scatterplot(x=col,y='Results',data=df,hue='Results',ax=axes[i][j]) 
    
    j = j + 1
    
    if j >= 2:
        j = 0
        if i < 5:
            i = i + 1

plt.show()
```
