```python
#importing the dataset from the url
import pandas as pd
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00571/hcvdat0.csv"
df = pd.read_csv(url)
```

```python
#removing unnecessary column
df = df.drop(["Unnamed: 0"],axis = 1)
```

```python
#preprocessing the dataset

#label encoding the categorical data 

from sklearn.preprocessing import LabelEncoder

#creating an instance for sex category encoding

enc1 = LabelEncoder()
#adding new categorized column to the dataframe
df["Gen_cat"] = enc1.fit_transform(df["Sex"])
df
```

```python
#encoding the target 

enc2 = LabelEncoder()

#addding encoded target value to the dataframe

df["Target"] = enc2.fit_transform(df["Category"])
```

```python
df.head()
```

```python
#removing non numerical columns

df_new = df.drop(columns = ["Category","Sex"],axis = 1)
df_new
```

```python
#checking for NaN values in the dataset

df_new.isnull().sum()
```

```python
#filling the NaN values

import numpy as np

df_new = df_new.fillna(method = "ffill")
df_new.isnull().sum()
```

```python
#dividing the dataset in to features and target

X = df_new.drop(["Target"],axis = 1)
Y = df_new["Target"]


```

```python
#splitting the dataset in to train and test splits

from sklearn.model_selection import train_test_split

X_tr, X_te, Y_tr, Y_te = train_test_split(X,Y,test_size = 0.2)
```

```python
#building SVM models 
from sklearn import svm

#linear kernel 
sv1 =  svm.SVC(kernel = "linear", C = 1) 

#trainig the linear kernel

sv1.fit(X_tr, Y_tr)
```

```python
#prediction on test data

Y_pre_l = sv1.predict(X_te)
```

```python
#evaluating the linear kernel performance

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print("Accuracy of the linear Kernel = ", accuracy_score(Y_te,Y_pre_l))
print("Precision of the linear Kernel = ", precision_score(Y_te,Y_pre_l,average = 'macro'))
print("Recall of the linear Kernel = ", recall_score(Y_te,Y_pre_l,average = 'macro'))
print("F1 score of the linear Kernel = ", f1_score(Y_te,Y_pre_l,average = "macro"))
```

```python
#polynomial kernel SVM 

sv2 = svm.SVC(kernel = "poly", C=1)

#training the polynomial kernel 

sv2.fit(X_tr,Y_tr)
```

```python
#prediction on test data
Y_pre_p = sv2.predict(X_te)
```

```python
#evaluating the performance in poly kernel

print("Accuracy of the polynomial Kernel = ", accuracy_score(Y_te,Y_pre_p))
print("Precision of the polynomial Kernel = ", precision_score(Y_te,Y_pre_p,average = 'macro'))
print("Recall of the polynomial Kernel = ", recall_score(Y_te,Y_pre_p,average = 'macro'))
print("F1 score of the polynomial Kernel = ", f1_score(Y_te,Y_pre_p,average = "macro"))
```

```python
#rbf kernel svm
sv3 = svm.SVC(kernel = 'rbf', C = 1)

#training the model

sv3.fit(X_tr, Y_tr)
```

```python
#prediction on test data
Y_pre_r = sv3.predict(X_te)
```

```python
#evaluating the performance in RBF kernel

print("Accuracy of the RBF Kernel = ", accuracy_score(Y_te,Y_pre_r))
print("Precision of the RBF Kernel = ", precision_score(Y_te,Y_pre_r,average = 'macro'))
print("Recall of the RBF Kernel = ", recall_score(Y_te,Y_pre_r,average = 'macro'))
print("F1 score of the RBF Kernel = ", f1_score(Y_te,Y_pre_r,average = "macro"))
```

```python
#implementing K-fold cross validation
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from numpy import mean,std
#for K = 10

kf10 = KFold(n_splits = 10, random_state=1,shuffle = True)

#Linear kernel model initialization

lc = svm.SVC(kernel = 'linear',C = 1)

#evaluating the linear kernel on K = 10 Kfold cross validation score
scores1 = cross_val_score(lc, X, Y, scoring='accuracy', cv=kf10, n_jobs=-1)
# report performance
print('Accuracy for linear kernel (K = 10): %.3f (%.3f)' % (mean(scores1), std(scores1)))




```

```python
#for K = 5
kf5 = KFold(n_splits = 5, random_state=1,shuffle = True)

#Linear kernel model initialization

lc2 = svm.SVC(kernel = 'linear',C = 1)

#evaluating the linear kernel on K = 5 Kfold cross validation score
scores2 = cross_val_score(lc2, X, Y, scoring='accuracy', cv=kf5, n_jobs=-1)
# report performance
print('Accuracy for linear kernel (K = 5): %.3f (%.3f)' % (mean(scores2), std(scores2)))
```

```python
#polynomial kernel initialization 

pk1 = svm.SVC(kernel = 'poly',C = 1)

#for k = 10
#evaluating the polynomial kernel on K = 10 Kfold cross validation score
scores3 = cross_val_score(pk1, X, Y, scoring='accuracy', cv=kf10, n_jobs=-1)
# report performance
print('Accuracy for polynomial kernel (K = 10): %.3f (%.3f)' % (mean(scores3), std(scores3)))
```

```python
#for k = 5
pk2 = svm.SVC(kernel = 'poly',C = 1)


#evaluating the polynomial kernel on K = 5 Kfold cross validation score
scores4 = cross_val_score(pk2, X, Y, scoring='accuracy', cv=kf5, n_jobs=-1)
# report performance
print('Accuracy for polynomial kernel (K = 5): %.3f (%.3f)' % (mean(scores4), std(scores4)))
```

```python
#rbf kernel

rbf1 = svm.SVC(kernel = 'rbf')
#for k = 10
#evaluating the rbf kernel on K = 10 Kfold cross validation score
scores5 = cross_val_score(rbf1, X, Y, scoring='accuracy', cv=kf10, n_jobs=-1)
# report performance
print('Accuracy for rbf kernel (K = 10): %.3f (%.3f)' % (mean(scores5), std(scores5)))
```

```python
#rbf kernel

rbf2 = svm.SVC(kernel = 'rbf')
#evaluating the rbf kernel on K = 5 Kfold cross validation score
scores6 = cross_val_score(rbf2, X, Y, scoring='accuracy', cv=kf5, n_jobs=-1)
# report performance
print('Accuracy for polynomial kernel (K = 5): %.3f (%.3f)' % (mean(scores6), std(scores6)))
```

```python
#creating a dataframe table for interpreting the results 

values = [[mean(scores1),mean(scores3),mean(scores5)],[mean(scores2),mean(scores4),mean(scores6)]]
data = pd.DataFrame(values, index = ["10 Fold Cross Validation Accuracy","5 Fold Cross Validation Accuracy"])
data.columns = ["Linear Kernel", "Polynomial Kernel", "RBF Kernel"]
data
```
